# Pipeline + StandardScaler + Cross-Validation: сравнение Lasso и Ridge

## Что происходит в этом ноутбуке

Исследуем, как **sklearn Pipeline**, **масштабирование признаков** и **кросс-валидация** работают вместе на примере датасета California Housing.

**Датасет:** California Housing — 20 640 блоков жилых домов в Калифорнии. Целевая переменная `MedHouseVal` — медианная стоимость дома (в сотнях тысяч $).

**Шаги:**
1. Загрузка и первичный осмотр данных.
2. Разделение на train/test (80/20).
3. Построение пайплайна `StandardScaler → Lasso` и подбор гиперпараметра `alpha` через `GridSearchCV` с 4-fold кросс-валидацией.
4. Аналогичный пайплайн `StandardScaler → Ridge` с подбором `alpha`.
5. Сравнение результатов двух моделей по метрике `neg_mean_squared_error`.

**Зачем Pipeline?** Чтобы масштабирование применялось корректно: scaler обучается **только на тренировочном фолде** внутри каждого сплита кросс-валидации, исключая утечку данных (data leakage).

```python
import numpy as np
```

```python
from sklearn.datasets import fetch_california_housing

df = fetch_california_housing(as_frame=True).frame

df.head(5)
```

```python
df.info()
```

```python
X = df.drop(columns='MedHouseVal')
Y = df['MedHouseVal']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y,
    test_size=0.2,
    shuffle=True,
)
```

```python
from sklearn.linear_model import Lasso
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV, KFold

pipe = make_pipeline(
    StandardScaler(),
    Lasso(max_iter=5000)
)

cv_lasso = GridSearchCV(
    pipe,
    param_grid={
        'lasso__alpha':np.linspace(0.00001, 1, num=3)
    },
    scoring='neg_mean_squared_error',
    cv=KFold(n_splits=4, shuffle=True, random_state=33),
    return_train_score=True
)

cv_lasso.fit(X_train, Y_train)

print(cv_lasso.best_params_)
print(cv_lasso.best_score_)
```

```python
from sklearn.linear_model import Ridge

pipe = make_pipeline(
    StandardScaler(),
    Ridge(max_iter=5000)
)

cv_lasso = GridSearchCV(
    pipe,
    param_grid={
        'ridge__alpha':np.linspace(0.00001, 10, num=3)
    },
    scoring='neg_mean_squared_error',
    cv=KFold(n_splits=4, shuffle=True, random_state=33),
    return_train_score=True
)

cv_lasso.fit(X_train, Y_train)

print(cv_lasso.best_params_)
print(cv_lasso.best_score_)
```

## Выводы

| Модель | Лучший `alpha` | Best MSE (neg) |
|--------|---------------|-----------------|
| **Lasso** (L1) | 0.500005 | **−0.958** |
| **Ridge** (L2) | 0.00001 | −2.171 |

1. **Lasso победил** — MSE ≈ 0.96 против ≈ 2.17 у Ridge. Lasso лучше справился с предсказанием стоимости жилья на этих данных.

2. **Lasso выбрал заметный `alpha`** (≈ 0.5), то есть активно зануляет часть коэффициентов. Это говорит о том, что не все 8 признаков одинаково полезны — Lasso выполняет неявный отбор фичей.

3. **Ridge выбрал минимальный `alpha`** (≈ 0), то есть регуляризация ему почти не помогает. При этом его ошибка существенно выше — модель не справляется с шумными/нерелевантными признаками.

4. **Pipeline защитил от data leakage**: StandardScaler внутри пайплайна гарантирует, что статистики (mean, std) вычисляются только по тренировочной части фолда.
