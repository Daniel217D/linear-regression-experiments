```python
import numpy as np
```

```python
from sklearn.datasets import fetch_california_housing

df = fetch_california_housing(as_frame=True).frame

df.head(5)
```

```python
df.info()
```

```python
X = df.drop(columns='MedHouseVal')
Y = df['MedHouseVal']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y,
    test_size=0.2,
    shuffle=True,
)
```

```python
from sklearn.linear_model import Lasso
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV, KFold

pipe = make_pipeline(
    StandardScaler(),
    Lasso(max_iter=5000)
)

cv_lasso = GridSearchCV(
    pipe,
    param_grid={
        'lasso__alpha':np.linspace(0.00001, 1, num=3)
    },
    scoring='neg_mean_squared_error',
    cv=KFold(n_splits=4, shuffle=True, random_state=33),
    return_train_score=True
)

cv_lasso.fit(X_train, Y_train)

print(cv_lasso.best_params_)
print(cv_lasso.best_score_)
```

```python
from sklearn.linear_model import Ridge

pipe = make_pipeline(
    StandardScaler(),
    Ridge(max_iter=5000)
)

cv_lasso = GridSearchCV(
    pipe,
    param_grid={
        'ridge__alpha':np.linspace(0.00001, 10, num=3)
    },
    scoring='neg_mean_squared_error',
    cv=KFold(n_splits=4, shuffle=True, random_state=33),
    return_train_score=True
)

cv_lasso.fit(X_train, Y_train)

print(cv_lasso.best_params_)
print(cv_lasso.best_score_)
```
